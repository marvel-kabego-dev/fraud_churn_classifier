```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/raw/synthetic_user_data.csv")
```

```python
X = df.drop(columns=['user_id', 'fraud_risk', 'churn_risk'])
```

```python
X = pd.get_dummies(X, columns=['payment_method'], drop_first=True)
```
